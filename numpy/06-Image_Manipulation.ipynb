```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = plt.imread(r'twobirds.jpg')
type(im)
```

```python
im.shape
```

```python
plt.imshow(im)
```

```python
R = im[:,:,0]
R
```

```python
fig, ax = plt.subplots(1,3)
ch = ['Red', 'Green', 'Blue']
for i,axi in enumerate(ax.flat):
    axi.imshow(im[:,:,i], plt.cm.gray)
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch[i], color=ch[i].lower())
    fig.set_size_inches(20,50)
```

```python
img2 = im.copy()
img2[:,:,2] = 200

fig2, ax2 = plt.subplots(1,3)
ch2 = ['Red', 'Green', 'Blue']
for i,axi in enumerate(ax2.flat):
    axi.imshow(img2[:,:,i], plt.cm.gray)
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch2[i], color=ch2[i].lower())
    fig2.set_size_inches(20,50)
```
